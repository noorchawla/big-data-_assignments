```python
import seaborn as sb
import pandas

df = pandas.read_csv("/content/sample_data/company_sales_data.csv")


graph = sb.lineplot(data=df, x="month_number", y="total_profit")
graph.set_xlabel("Months")
graph.set_ylabel("Total Profit")
graph.set_title("Total Profit per month")
```

```python
columns = list(filter(lambda x: x not in ["month_number", "total_units", "total_profit"], df.keys()))
print(columns)
for col in columns:
  sb.lineplot(data=df, x="month_number", y=col, label=col).set(title='Items Sales Per Month', xlabel='Months', ylabel='Units Sold')
```

```python
import plotly.express as ply
df=ply.data.iris()


fig = ply.scatter(data_frame=df,x='sepal_width',y='sepal_length', title="Sepal Length vs Width")
fig.update_layout(xaxis_title="Sepal Width", yaxis_title="Sepal Length")
fig.show()
```

```python
import plotly.express as ply
df=ply.data.iris()


fig = ply.bar(data_frame=df,x='sepal_width',y='sepal_length', title="Sepal Length vs Width")
fig.update_layout(xaxis_title="Sepal Width", yaxis_title="Sepal Length")
fig.show()
```
