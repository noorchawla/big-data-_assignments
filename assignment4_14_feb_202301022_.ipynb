```python
import pandas as pd
import matplotlib.pyplot as plt

sales_data = pd.read_csv('company_sales_data.csv')
print(sales_data.head())

plt.figure(figsize=(10, 6))
plt.plot(sales_data['month_number'], sales_data['total_profit'], marker='o', color='b', label='Total Profit')
plt.title('Total Profit of All Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.legend()
plt.show()
```

```python

products = sales_data.columns[1:7]
plt.figure(figsize=(12, 8))
for product in products:
    plt.plot(sales_data['month_number'], sales_data[product], marker='o', label=product)

plt.title('Product Sales Data Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sales_data.drop(['month_number', 'total_units', 'total_profit'], axis=1).sum().plot(kind='bar', color='skyblue')
plt.title('Bar Chart for All Features/Attributes')
plt.xlabel('Feature/Attribute')
plt.ylabel('Sum')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
attributes_to_plot = sales_data.columns[1:5]
colors = ['red', 'green', 'blue', 'orange']

for i, attribute in enumerate(attributes_to_plot):
    plt.scatter(sales_data['month_number'], sales_data[attribute], color=colors[i], label=attribute)

plt.title('Scatter Plot for 4 Attributes')
plt.xlabel('Month')
plt.ylabel('Attribute Value')
plt.legend()
plt.show()
```
